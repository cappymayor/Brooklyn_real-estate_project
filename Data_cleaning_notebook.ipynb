After checking the null values above, we can see the geographical region has 7 missing values which weintentionally made it this way because of unclarity with one of the neighborhood(JAMAICA BAY). Politically, Jamaica bay seems not to be a neigborhood in Brooklyn, it is primarily divided between the boroughs of Brooklyn and Queens in New York City, with a small part touching Nassau County, so on this basis we decided not to fill it with geographical region we are not certain of, hence we decide we will drop it.

```python
# checking where neighborhood is jamaica bay so we have a look of how many rows
subset_df.loc[subset_df["geographical_region"].isnull()]
```

```python
#its not a lot of row, so we drop this rows
subset_df = subset_df.dropna()
```

```python
subset_df.isnull().sum()
```

Good that we no longer have null values, however in our dataframe under the sale_price columns, there are properties sale price that are less than $1000 and many more, so we decided to set a pricing threshold of 100000 above so we can use this in our analyis.
